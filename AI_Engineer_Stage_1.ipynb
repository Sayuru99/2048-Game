```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
image = img.imread('Angela-yu.jpg')
fig,axes =  plt.subplots(2,3,figsize = (7,7))

axes[0][0].imshow(image[:,:,0],cmap='Reds')
axes[0][1].imshow(image[:,:,1],cmap='Greens')
axes[0][2].imshow(image[:,:,2],cmap='Blues')
axes[1][0].imshow(image[:,:,0],cmap='gray')
axes[1][1].imshow(image[:,:,1],cmap='gray')
axes[1][2].imshow(image[:,:,2],cmap='gray')

plt.tight_layout()

plt.show() # display all the plots
```
